```python
import pandas as pd
from sklearn.cluster import MeanShift
```

```python
df = pd.read_excel('Тестовое.xlsx', sheet_name='Player_Groups')
```

```python
df.head()
```

```python
df.dropna(inplace=True, axis=1) # уберем столбцы, где отсутствуют значения
```

```python
X = df.drop(['player'], axis=1)
```

```python
# применим алгоритм сдвига среднего значения кластеризации по имеющимся признакам
ms = MeanShift()
ms.fit(X)
labels = ms.labels_
```

```python
labels
```

```python
result_df = df[['player']]
```

```python
result_df = result_df.join(pd.DataFrame(labels, columns=['groupNum']))
```
